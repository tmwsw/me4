### Инициализация нужных данных

```python
df = pd.read_csv('Данные в нужном формате.csv')

df
```

### Удаление пропусков после разметки данных

```python
df_clear = df.dropna(how='all')

df_clear
```

### Удаление ненужных столбцов

```python
df = df_clear.drop(columns=['annotator', 'annotation_id', 'created_at', 'lead_time', 'updated_at'], axis=1)

df
```

### Цикл для заполенения столбцов категориями

```python
categories = [
    'Вопрос решен',
    'Качество выполненной работы',
    'Компетентность работника в общение',
    'Не удовлетворен',
    'Скорость выполненной работы',
    'Скорость ответа работника после запроса',
    'Не понравился результат',
    'Не понравилось качество'
]

# Заполняем 1 или 0 сразу по категориям, без создания промежуточной колонки
for cat in categories:
    df[cat] = df['taxonomy'].apply(
        lambda x: int(
            cat in [
                subitem[0]
                for item in ast.literal_eval(x)
                if 'taxonomy' in item
                for subitem in item['taxonomy']
            ]
        ) if x else 0
    )
```

### Изменение индекса и удаление ненужного столбца

```python
df.reset_index(drop=True, inplace=True)
df.index += 1

df = df.drop('taxonomy', axis=1)
```

### Мердж колонок так как они имеют общий смысл и малое количество данных

```python
df['Скорость выполненной работы'] = df[['Скорость выполненной работы', 'Скорость ответа работника после запроса']].max(axis=1)
df = df.drop('Скорость ответа работника после запроса', axis=1)
df
```

### Вывод количества отмеченных классов в колонках

```python
columns = [
    'Вопрос решен',
    'Качество выполненной работы',
    'Компетентность работника в общение',
    'Не удовлетворен',
    'Скорость выполненной работы',
    'Не понравился результат',
    'Не понравилось качество'
]

for col in columns:
    print(f"{col}: {df[col].sum()}")
```
